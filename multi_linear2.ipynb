```python
# Eksik verisi olan satırları tamamen sil
data = data.dropna()
```

```python
# DataFrame'deki en yüksek 5 mutlak korelasyonu bulur
top5 = data.corr().abs().unstack().sort_values(ascending=False)  # Tüm değişken çiftleri arasındaki mutlak korelasyonları sırala
top5 = top5[top5 < 1].drop_duplicates().head(5)  # Kendisiyle olan korelasyonları çıkar ve en yüksek 5'ini al
print(top5)  # Sonuçları yazdır
# 0.75 üstü olanları yazdır
print("🔍 0.75'in üzerindeki korelasyon çiftleri:\n")
for (col1, col2), val in top5.items():
    if val > 0.75:
        print(f"{col1} ↔ {col2} : {val:.3f}")

```

```python
data.columns = data.columns.str.strip()
x = data.drop(["DIS", "RAD", "MEDV"], axis=1)
y = data["MEDV"]
```

```python
from sklearn.model_selection import train_test_split
# Veriyi böl
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
# Modeli kur
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
y_pred = model.predict(x_test)
```

```python
print("R² Skoru:", r2_score(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.7, color="royalblue", label="Tahminler")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color="red", linestyle="--", label="Mükemmel Doğruluk Çizgisi")
plt.xlabel("Gerçek MEDV Değerleri")
plt.ylabel("Tahmin Edilen MEDV Değerleri")
plt.title("Gerçek vs Tahmin MEDV Fiyatları")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
